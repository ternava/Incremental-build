```python
import os
import pandas as pnd
import graphviz
```

```python
def dict_to_tuples(d):
    return [(d["inc"][i], d["cle"][i], d["time"][i]) for i in range(len(d["inc"]))]        
```

```python
def is_present(graph, value):
    for k in graph.body:
        if value in k:
            return True
    return False
```

```python
def boex_graph(l):
    g = graphviz.Digraph()
    for t in l:
        ic, cl, ti = str(t[0]), str(t[1]), str(t[2])
        if not is_present(g, ic):
            g.node(ic, ic)
        if not is_present(g, cl):
            g.node(cl, cl)
        g.edge(cl, ic, ti)
    return g
```

```python
def csv_to_g(csv):
    return boex_graph(dict_to_tuples(csv.to_dict()))
```

```python
graphs = dict()
dire = "order"
for k in os.listdir(dire):
    file = "{}/{}".format(dire, k)
    dict_entry = k.split(".")[0]
    graphs[dict_entry] = csv_to_g(\
                                  pnd.read_csv(file))
```
