## Configurations

```python
import ast
```

```python
content = None
with open("config_data", 'r') as data:
    content = data.read()
```

```python
cdict = ast.literal_eval(content)
```

```python
all_cpp_options_set = set()
for conf in cdict:
    for opt in cdict[conf]:
        all_cpp_options_set.add(opt)
```

```python
all_cpp_options = list(all_cpp_options_set)
```

```python
ttbl = {"cpp_val" : all_cpp_options}
```

```python
lall = len(all_cpp_options)
for conf in cdict:
    ttbl[conf] = []
    for i in range(lall):
        if ttbl["cpp_val"][i] in cdict[conf]:
            ttbl[conf].append(cdict[conf][ttbl["cpp_val"][i]])
        else:
            ttbl[conf].append(-1)
```

```python
df = pnd.DataFrame.from_dict(ttbl)
```
