## I-3. Configurations

```python
import ast
```

```python
def dict_file_to_dict(dict_file):
    content = None
    with open(dict_file, 'r') as data:
        content = data.read()
    return ast.literal_eval(content)
```

```python
def get_all_cpp_options(configs_dict):
    all_cpp_options_set = set()
    for conf in configs_dict:
        for opt in configs_dict[conf]:
            all_cpp_options_set.add(opt)
    return list(all_cpp_options_set)
```

```python
def dict_to_df_struct(cdict):
    all_cpp_options = get_all_cpp_options(cdict)
    ttbl = {"cpp_val" : all_cpp_options}
    for conf in cdict:
        ttbl[conf] = []
        for i in range(len(all_cpp_options)):
            if ttbl["cpp_val"][i] in cdict[conf]:
                ttbl[conf].append(cdict[conf][ttbl["cpp_val"][i]])
            else:
                ttbl[conf].append(-1)
    return ttbl
```

```python
def to_df(cfile):
    cdict = dict_file_to_dict(cfile)
    dres = dict_to_df_struct(cdict)
    return pnd.DataFrame.from_dict(dres)
```

```python
to_df("config_data")
```
