# Make clean build time

```python
import pandas as pd
import numpy as np
```

```python
# It shows the clean Build Time and the executable's Binary Size 
# for all built system's configurations

path = '/src/data/rez_sample_03ALL/'
df = pd.read_csv(path + 'buildtime_dc2.csv')
df.columns
df = df.rename(columns = {"bt_real": "Build Time", "BinarySize": "Binary Size"})
df
```

```python
# From the previous row data, here are extracted only 
# the clean Build Time and Binary Size data 

# The "BranchC" column show the name of built configurations.
# e.g., 00005 is the /configuration/sample-quicktest/00003.config 

df[['Br0', 'BranchC']] = df.Branch.str.split("-",expand=True)

df_setindex = df.set_index("BranchC")
df = df_setindex[['Build Time', 'Binary Size']]

# The index is reset to the clean build configuration
# Saved these data to a new CSV file

dx = df.sort_index()
dx.to_csv(path + 'clean-bt.csv')
dx
```
