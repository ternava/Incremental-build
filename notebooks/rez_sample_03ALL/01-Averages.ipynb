Average of clean build for sample_03

```python
import numpy as np
import pandas as pd

# Find the average of clean Built Time (column "bt_real")
# and the average of Binary Size in both executions 

path = '/src/data/'
df1 = pd.read_csv(path + 'rez_sample_03I/buildtime_dc2.csv')
df2 = pd.read_csv(path + 'rez_sample_03II/buildtime_dc2.csv')

df = pd.concat([df1, df2])
df = df.groupby(['Branch', 'Option']).agg({
    "bt_real":np.mean,
    "BinarySize": np.mean
}).reset_index()

# Saved their average to a new CSV file

df.to_csv(path + 'rez_sample_03ALL/buildtime_dc2.csv')
df
```
