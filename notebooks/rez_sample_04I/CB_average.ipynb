Average of clean build for sample_03

```python
import numpy as np
import pandas as pd

path = '/src/data/'
df1 = pd.read_csv(path + 'rez_sample_04I/buildtime_dc2.csv')
df2 = pd.read_csv(path + 'rez_sample_04II/buildtime_dc2.csv')

df = pd.concat([df1, df2])
df = df.groupby(['Branch', 'Option']).agg({
    "bt_real":np.mean,
    "BinarySize": np.mean
}) #.reset_index()

# Saved to a new csv file
#df = df.sort_index()
df.to_csv(path + 'rez_sample_04ALL/buildtime_dc2.csv')

df
```
